```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("G:/Coding/Course related/ML Resource-20250103T081228Z-001/ML Resource/height-weight.csv")
```

```python
df
```

```python
plt.scatter (df.Weight, df.Height)
plt.xlabel('Weight')
plt.ylabel('Height')
```

```python
#problem statement : i want to predict the height based on weight
```

```python
X = df[['Weight']]
y = df[['Height']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
print("The slope or coeff of model is ",model.coef_)
print("The intercept of model is ",model.intercept_)
```

```python
model.predict(X_test)
```

```python
# this plot is based on prediction of train data
plt.scatter(X_train,y_train)
plt.plot(X_train,model.predict(X_train), color = 'green')
```

```python
#this plot is based on prediction of test data
plt.scatter (X_test, y_test, color = 'black')
plt.plot(X_test, model.predict(X_test), color = 'cyan')
```

```python
mean_absolute_error (y_test, model.predict(X_test))
```

```python
mean_squared_error (y_test, model.predict(X_test))
```

```python
rsq = r2_score(y_test, model.predict(X_test))
```
