```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("G:/Coding/Datasets/bank-additional-full.csv", sep = ';')
```

```python
df
```

```python
no = df[df['y'] == 'no']
no
```

```python
yes = df[df['y'] == 'yes']
yes
```

```python
#since we have 36000+ rows for no and 4500+ rows for yes, we will take 4500 rows from no to make the data balanced to avoid bias
no_sample = no.sample(4640, random_state = 42)
```

```python
new_df = pd.concat([no_sample,yes])
new_df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
data_df = new_df.apply(encoder.fit_transform)
```

```python
data_df
```

```python
data_df['y'].value_counts()
```

```python
#making value within the range of 0 and 1
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
normalized_data = scaler.fit_transform(data_df)
normalized_data
```

```python
corr = data_df.corr()
plt.figure(figsize = (20,20))
sns.heatmap (corr, cmap = 'RdBu', annot = True)
plt.show()
```

```python
x = normalized_data[:,0:20]
y = normalized_data[:,20]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train,y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```
