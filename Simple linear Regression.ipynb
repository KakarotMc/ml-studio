```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
#problem statement : predict the diabetes progression after 1 year based on the given features
```

```python
df = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
```

```python
df['target'] = diabetes.target
```

```python
df
```

```python
#after this EDA, data cleaning, Data Preparation, Feature engineering but in our case everything is done before
```

```python
X = df[['bmi']] #independent variable
y = df[['target']] #dependent variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color = 'black', label = 'Actual')
plt.plot(X_test, y_pred, color = 'cyan',linewidth = 3, label = 'Predicted Linear regression')
plt.xlabel('BMI')
plt.ylabel('Diabetes Progression')
plt.title('Linear Regression on Diabetes Progression')
plt.legend()
plt.show()
```

```python
mean_squared_error(y_test, y_pred)
```
